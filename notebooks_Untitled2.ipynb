```python
def get_odd_numbers():
    odd_list = []
    for i in range(1, 26):
        if i % 2 != 0:
            odd_list.append(i)
    return odd_list

print(get_odd_numbers())
```

```python
# *args example: Summing any number of integers
def sum_all(*args):
    return sum(args)

# **kwargs example: Printing user profile data
def build_profile(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print(sum_all(10, 20, 30))
build_profile(name="Gemini", role="AI", version="3.0")
```

```python
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1
```

```python
def prime_generator():
    num = 2
    while num < 1000:
        is_prime = True
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            yield num
        num += 1

gen = prime_generator()
for _ in range(20):
    print(next(gen), end=" ")
```

```python
def is_palindrome(num):
    temp = num
    reverse = 0
    while temp > 0:
        digit = temp % 10
        reverse = reverse * 10 + digit
        temp //= 10
    return num == reverse

number = 121
print(f"Is {number} a palindrome? {is_palindrome(number)}")
```
